```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/julia/data/gradients/results/regions/finest_regions.csv')
df = df.drop(columns=['Unnamed: 0', 'comment'])
df = df.dropna()
df = df.reindex()
```

```python
sns.set_context('talk')
fig = plt.figure(figsize=(10,10))

plt.scatter(df.loc[df['cluster']=='prefrontal', 'gradient 2'], 
            df.loc[df['cluster']=='prefrontal', 'gradient 4'],
           color='tab:red', label='prefrontal')
plt.scatter(df.loc[df['cluster']=='lateral', 'gradient 2'], 
            df.loc[df['cluster']=='lateral', 'gradient 4'],
           color='gold', label='lateral')
plt.scatter(df.loc[df['cluster']=='medial', 'gradient 2'], 
            df.loc[df['cluster']=='medial', 'gradient 4'],
           color='darkblue', label='medial')
plt.scatter(df.loc[df['cluster']=='visual', 'gradient 2'], 
            df.loc[df['cluster']=='visual', 'gradient 4'],
           color='cornflowerblue', label='visual')
plt.scatter(df.loc[df['cluster']=='auditory', 'gradient 2'], 
            df.loc[df['cluster']=='auditory', 'gradient 4'],
           color='slateblue', label='auditory')
plt.scatter(df.loc[df['cluster']=='somatomotor', 'gradient 2'], 
            df.loc[df['cluster']=='somatomotor', 'gradient 4'],
           color='tab:orange', label='somatomotor')
plt.ylim(-0.025, 0.025)
plt.xlim(-0.035, 0.035)
plt.xlabel('Gradient 2')
plt.ylabel('Gradient 4')
plt.legend(frameon=False)
sns.despine()
```
